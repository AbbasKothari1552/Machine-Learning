# Fetch California Housing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
```

```python
data
```

```python
df = pd.DataFrame(data.data)
```

```python
df
```

```python
df.columns = data.feature_names
```

```python
df
```

```python
# independent & dependent features
X = df
y = data.target
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_test
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## apply ridge regression

```python
# import Ridge & GridSearchCV it is for selecting the best alpha value (lambda) from the given
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge_reg = Ridge()
```

```python
# Possible Parameters for the value of alpha.
parameters = {'alpha' : [1,2,5,9,10,15,18,20,27,35,30,37,40,50,55]}
```

```python
ridge_cv = GridSearchCV(ridge_reg,parameters,scoring='neg_mean_squared_error',cv=5)
```

```python
ridge_cv.fit(X_train,y_train)
```

```python
print(ridge_cv.best_params_)
```

```python
print(ridge_cv.best_score_)
```

```python
# predict the test model
ridge_pred = ridge_cv.predict(X_test)
```

```python
import seaborn as sns
sns.displot(ridge_pred-y_test,kind = 'kde')
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,ridge_pred)
```

```python
score
```

# Lasso Regression

```python
# import Lasso
from sklearn.linear_model import Lasso
```

```python
lasso_reg = Lasso()
```

```python
# Possible Parameters for the value of alpha.
parameters = {'alpha' : [1,2,5,9,10,15,18,20,27,35,30,37,40,50,55]}
```

```python
lasso_cv = GridSearchCV(lasso_reg,parameters,scoring='neg_mean_squared_error',cv=5)
```

```python
lasso_cv.fit(X_train,y_train)
```

```python
print(lasso_cv.best_params_)
```

```python
print(lasso_cv.best_score_)
```

```python
# predict
lasso_pred = lasso_cv.predict(X_test)
```

```python
sns.displot(lasso_pred-y_test,kind='kde')
```

```python
# r2_score
from sklearn.metrics import r2_score
score = r2_score(y_test,lasso_pred)
```
